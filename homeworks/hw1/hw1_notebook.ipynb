```
from deepul.hw1_helper import *

# Copied form the solutions of the homework, this is a general trainer

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torch.utils.data as data
import matplotlib.pyplot as plt

quiet = True

def train(model, train_loader, optimizer, epoch, grad_clip=None):
  model.train()
  
  train_losses = []
  for x in train_loader:
    x = x.cuda().contiguous()
    loss = model.loss(x)
    optimizer.zero_grad()
    loss.backward()
    if grad_clip:
      torch.nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
    optimizer.step()
    train_losses.append(loss.item())
  return train_losses

def eval_loss(model, data_loader):
  model.eval()
  total_loss = 0
  with torch.no_grad():
    for x in data_loader:
      x = x.cuda().contiguous()
      loss = model.loss(x)
      total_loss += loss * x.shape[0]
    avg_loss = total_loss / len(data_loader.dataset)

  return avg_loss.item()


def train_epochs(model, train_loader, test_loader, train_args):
  epochs, lr = train_args['epochs'], train_args['lr']
  grad_clip = train_args.get('grad_clip', None)
  print(model)
  optimizer = optim.Adam(model.parameters(), lr=lr)

  train_losses = []
  test_losses = [eval_loss(model, test_loader)]
  for epoch in range(epochs):
    model.train()
    train_losses.extend(train(model, train_loader, optimizer, epoch, grad_clip))
    test_loss = eval_loss(model, test_loader)
    test_losses.append(test_loss)
    if not quiet:
      print(f'Epoch {epoch}, Test loss {test_loss:.4f}')

  return train_losses, test_losses
```

```
class HistogramGenerator(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.theta = nn.Parameter(torch.zeros(1, d), requires_grad=True)
    
    def loss(self, x):
        logits = self.theta.repeat(x.shape[0], 1)
        # Take the negative log likelihood of the softmax, that is the same as 
        # taking the cross entropy loss
        return F.cross_entropy(logits, x.long())

    def get_distribution(self):
        distribution = F.softmax(self.theta.squeeze(), dim=0)
        return distribution.detach().cpu().numpy()
```

## Part (b) Fitting Discretized Mixture of Logistics

Let us model $p_\theta(x)$ as a **discretized** mixture of 4 logistics such that $p_\theta(x) = \sum_{i=1}^4 \pi_i[\sigma((x+0.5 - \mu_i)/s_i) - \sigma((x-0.5-\mu_i)/s_i)]$

For the edge case of when $x = 0$, we replace $x-0.5$ by $-\infty$, and for $x = d-1$, we replace $x+0.5$ by $\infty$.

You may find the [PixelCNN++](https://arxiv.org/abs/1701.05517) helpful for more information on discretized mixture of logistics.

**Provide the same set of corresponding deliverables as part (a)**

### Solution
Fill out the function below and return the necessary arguments. Feel free to create more cells if need be.

```
class MixtureOfLogistics(nn.Module):
    def __init__(self, d, n_mixture=4):
        super().__init__()
        self.d = d
        self.n_mixture = n_mixture 

        # logits later pass through softmax to get mixture weights
        self.logits = nn.Parameter(torch.zeros(n_mixture), requires_grad=True)

        # Initialize evenly spaced means and random standard deviations
        # All sizes 1 x n_mixture after unsqueezing
        init_means = torch.arange(0, n_mixture).float() / (n_mixture - 1) * d 
        self.means = nn.Parameter(init_means, requires_grad=True)
        self.log_scales = nn.Parameter(
            torch.rand(self.n_mixture), requires_grad=True)
    
    def forward(self, x):
        
        # batch x n_mixture
        d = self.d
        x = x.float().unsqueeze(1).repeat(1, self.n_mixture)
        
        # Unsqueeze here
        means, log_scales = self.means.unsqueeze(0), self.log_scales.unsqueeze(0)
     
        # log_scales is easier to optimize than the scale itself
        inv_scales = torch.exp(-log_scales)
        
        plus_bell = torch.sigmoid(inv_scales * (x + 0.5 - means))
        minus_bell = torch.sigmoid(inv_scales * (x - 0.5 - means))
        logistics = plus_bell - minus_bell

        log_logistics = torch.log(torch.clamp(logistics, min=1e-12))
        
        # Logarithm of the edge case boundary [x = 0]
        boundary_plus = torch.sigmoid(inv_scales * (0.5 - means))
        log_boundary_plus = torch.log(torch.clamp(boundary_plus, min=1e-12))

        # Logarithm of the edge case boundary [x = d-1]
        boundary_min = 1 - torch.sigmoid(inv_scales * ((d-1) - 0.5 - means))
        log_boundary_min = torch.log(torch.clamp(boundary_min, min=1e-12))
        
        # Cover edge cases
        x_log_probs = torch.where(x < 0.001,        log_boundary_plus,
                      torch.where(x > d - 1 - 1e-3, log_boundary_min,
                                                    log_logistics))

        pi_log = F.log_softmax(self.logits, dim=0).unsqueeze(0)
        
        # multiplication is sum in log probs
        log_prob = x_log_probs + pi_log 

        # Sum log_probs and later take the log, the output of this
        # network is a logprob
        return torch.logsumexp(log_prob, dim=1) 

    def loss(self, x):
        # Negative log likelihood (output of net is a logprob) and mean over 
        # batch
        return -torch.mean(self(x))

    def get_distribution(self):
        with torch.no_grad():
            x = torch.arange(self.d).float().cuda()
            distribution = self(x).exp()
        return distribution.detach().cpu().numpy()
```

```
def q1_b(train_data, test_data, d, dset_id):
    """
    train_data: An (n_train,) numpy array of integers in {0, ..., d-1}
    test_data: An (n_test,) numpy array of integers in {0, .., d-1}
    d: The number of possible discrete values for random variable x
    dset_id: An identifying number of which dataset is given (1 or 2). Most likely
             used to set different hyperparameters for different datasets

    Returns
    - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
    - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
    - a numpy array of size (d,) of model probabilities
    """
    
    model = MixtureOfLogistics(d, n_mixture=4).cuda()
    train_loader = data.DataLoader(train_data, batch_size=128, shuffle=True)
    test_loader = data.DataLoader(test_data, batch_size=128)

    train_losses, test_losses = train_epochs(model, train_loader, test_loader, 
                                             {"epochs": 20, "lr": 0.1})

    distribution = model.get_distribution()
    return train_losses, test_losses, distribution
```

```
def to_one_hot(labels, d):
    one_hot = torch.FloatTensor(labels.shape[0], d).cuda()
    one_hot.zero_()
    one_hot.scatter_(1, labels.unsqueeze(1), 1)
    return one_hot

# From Karpathys pytorch MADE implementation
class MaskedLinear(nn.Linear):
    """ same as Linear except has a configurable mask on the weights """
    
    def __init__(self, in_features, out_features, bias=True):
        super().__init__(in_features, out_features, bias)        
        self.register_buffer('mask', torch.ones(out_features, in_features))
        
    def set_mask(self, mask):
        self.mask.data.copy_(torch.from_numpy(mask.astype(np.uint8).T))
        
    def forward(self, input):
        return F.linear(input, self.mask * self.weight, self.bias)

class MaskedMLP(nn.Module):
    """ Masked MLP with ReLU activation functions, the mask can be changed
    by accessing the individual MasedLinear layer "set_mask" function
    """
    def __init__(self, n_input, n_output, hidden_sizes=(100, 100)):
        super().__init__()
        self.n_input = n_input
        self.n_output = n_output
        self.hidden_sizes = hidden_sizes

        # Get sizes and construct the network
        layers = []
        sizes = [n_input] + hidden_sizes + [n_output]
        for n_in, n_out in zip(sizes, sizes[1:]):
            layers.extend([
                MaskedLinear(n_in, n_out),
                nn.ReLU()    
            ])

        layers.pop() 
        self.layers = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.layers(x)

class MADE(nn.Module):
    def __init__(self, input_shape, d, hidden_sizes=[100, 100], one_hot_encoding=False):
        super().__init__()
        self.d = d
        self.hidden_sizes = hidden_sizes

        # MADE is for discrete values so we need to multiply our number of inputs
        # by the possible number of discrete values in our outpus.
        self.input_shape = input_shape
        self.n_in = np.prod(input_shape)

        # The output is the number of random variables times the number of possible
        # discrete values
        self.n_out = self.n_in * d
        self.one_hot_encoding = one_hot_encoding

        # Define network, choose if we are gonna use one hot encoding or not 
        input_size = self.n_in * d if one_hot_encoding else self.n_in
        self.backbone = MaskedMLP(input_size, self.n_out, hidden_sizes)
        
        # Create the masks for the MultiLayerPerceptron
        self.max_connections = {}
        self.create_mask()
    
    def create_mask(self):
        n_hidden = len(self.hidden_sizes)
        
        # Max connections of input layer
        self.max_connections[-1] = np.arange(self.n_in)
        
        # Randomly sample connections for hidden layers, satisfying the property
        # that unit k of layer l has at least the same maximum number of 
        # connections as unit k of layer l-1
        for l in range(n_hidden):
            self.max_connections[l] = np.random.randint(
                self.max_connections[l-1].min(), 
                self.n_in - 1,
                size=self.hidden_sizes[l])
    
        # Build masks for every layer (strictly lower dimensional matrices)
        masks = [self.max_connections[l - 1][:, None] <= self.max_connections[l][None, :] 
                 for l in range(n_hidden)]

        # Last layer should make possible p(x_i | x_<i) 
        masks.append(self.max_connections[l - 1][:, None] < self.max_connections[-1][None, :])
        
        # Last layer should map to discrete values for the pixel
        masks[-1] = masks[-1].repeat(self.d, axis=1)
        # If one hot encoding first layer should also do this
        if self.one_hot_encoding:
            masks[0] = masks[0].repeat(self.d, axis=0)
        
        masked_layers= [l for l in self.backbone.modules() if isinstance(l, MaskedLinear)]
        for l, m in zip(masked_layers, masks):
            l.set_mask(m)

    def forward(self, x):
        batch_size = x.shape[0]
        if self.one_hot_encoding:
            x = F.one_hot(x, self.d).float()
            x = x.view(batch_size, -1)
        else:
            x = x.float().view(batch_size, self.n_in)
        logits = self.backbone(x)
        # Shape the output to have the joint probability of (x1, x2) with
        # d discrete values -> [B, D, N_in]
        logits = logits.view(batch_size, self.n_in, self.d).permute(0, 2, 1)
        logits = logits.contiguous().view(batch_size, self.d, *self.input_shape)
        return logits 

    def loss(self, x):
        # Negative log likelihood (output of net is a logprob) and mean over 
        # batch
        return F.cross_entropy(self(x), x.long())

    def get_distribution(self):
        assert self.input_shape == (2,)
        all_pixels = np.mgrid[0:self.d, 0:self.d].reshape(2, self.d ** 2).T
        all_pixels = torch.LongTensor(all_pixels).cuda()

        output = self(all_pixels)

        log_probs = F.log_softmax(output, dim=1)
        # Gather values along D axis using all_pixels(B, 1, 2) as indexes
        log_dist = torch.gather(log_probs, 1, all_pixels.unsqueeze(1)).squeeze(1)
        # Get prod of p(x_1) * p(x_2 | x_1) = p(x_1, x_2)
        log_dist = log_dist.sum(dim=1)
        # And now get the probability of the image
        image_prob = log_dist.exp().view(self.d, self.d)
        return image_prob.detach().cpu().numpy()
```

```
def q2_a(train_data, test_data, d, dset_id):
  """
  train_data: An (n_train, 2) numpy array of integers in {0, ..., d-1}
  test_data: An (n_test, 2) numpy array of integers in {0, .., d-1}
  d: The number of possible discrete values for each random variable x1 and x2
  dset_id: An identifying number of which dataset is given (1 or 2). Most likely
           used to set different hyperparameters for different datasets

  Returns
  - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
  - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
  - a numpy array of size (d, d) of probabilities (the learned joint distribution)
  """
  
  model = MADE((2,), d, hidden_sizes=[256, 256], one_hot_encoding=True).cuda()
  train_loader = data.DataLoader(train_data, batch_size=128, shuffle=True)
  test_loader = data.DataLoader(test_data, batch_size=128)
  train_losses, test_losses = train_epochs(model, train_loader, test_loader, 
                                           dict(epochs=20, lr=2e-3))
  distribution = model.get_distribution()

  return train_losses, test_losses, distribution
```
